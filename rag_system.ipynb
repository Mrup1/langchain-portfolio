```python
#necessary libaries
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint , HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

```python
#here we are fetching the transcript from the yt vedio

video_id = "ShYKkPPhOoc" # id of the youtube vedio
try:
    transcript_list = YouTubeTranscriptApi.get_transcript(video_id, languages=["en"])#choose language of the transcript

    #printing the orignal list containg text , time and duration
    print(transcript_list)
    # flatting the transcript
    transcript = " ".join(item["text"] for item in transcript_list)
    print(transcript)

except TranscriptsDisabled:
    print("No captions available for this video.")
```

```python
#now we have to split the transcript in to smaller chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.create_documents([transcript])
```

```python
print(len(chunks))
```

```python
print(chunks[0].page_content)
```

```python
model_name = "sentence-transformers/all-MiniLM-L6-v2"

embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs={'device': 'cpu'},
    encode_kwargs={'normalize_embeddings': True}
)
vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
#viewing the embeddings
vector_store.index_to_docstore_id
```

```python
#veiwing a document
vector_store.get_by_ids(['d12c3585-b977-4bf5-a8a1-9fc35463cf4b'])
```

```python
#now making a retriver that will retrive the relevent document from the documents that we have created in previous step
retriever = vector_store.as_retriever(search_type="mmr", search_kwargs={"k": 3})
```

```python
#making a query
retriever.invoke("what is this vedio for")
```

```python
llm_huggingface = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    temperature=0.1,
    max_new_tokens=100,
    timeout=120
)

model = ChatHuggingFace(llm=llm_huggingface)
```

```python
prompt = PromptTemplate(
    template="""
      You are a helpful ai assistant.
      answer only from the provided transcript context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
question = "how to spend our 20 in ai era"
retrived_doc = retriever.invoke(question , k = 3)
```

```python
retrived_doc
```

```python
context = "  ".join(retrived_doc.page_content for retrived_doc in retrived_doc)
```

```python
context
```

```python
final_prompt = prompt.invoke({'context': context , 'question': question })
```

```python
final_prompt
```

```python
result = model.invoke(final_prompt)
```

```python
print(result.content)
```

doing this through chains

```python
from langchain_core.runnables import RunnableParallel , RunnableSequence , RunnablePassthrough , RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_doc(retrived_doc):
  context_doc = " ".join(retrived_doc.page_content for retrived_doc in retrived_doc)
  return context_doc
```

```python
parallel_chain = RunnableParallel({
    'context' : retriever | RunnableLambda(format_doc),
    'question' : RunnablePassthrough()
})
```

```python
parallel_chain.invoke("who is speaking in the vedio")
```

```python
parser = StrOutputParser()
```

```python
final_chain = parallel_chain | prompt | model | parser
```

```python
final_chain.invoke('summarize the vedio')
```
